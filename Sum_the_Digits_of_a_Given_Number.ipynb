```python
#Using modulo (%)
n=int(input())
sum = 0

while n > 0:
    sum += n % 10  # extract last digit
    n //= 10       # remove last digit

print(sum)
```

```python
#Using recursion
def fun(n):
    if n == 0:
        return 0  # base case
    return (n % 10) + fun(n // 10)

n = int(input())
print(fun(n))
```
